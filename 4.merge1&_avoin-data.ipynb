```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd

dx = pd.read_csv('merge/merge1_koordinaatit&kunnat.csv', delimiter=',')
dy = pd.read_csv('raw/_avoin-data-suomen-tyokyvyttomyyselakkeensaajat-sairauden-mukaan.csv', delimiter=',')
```

```python
dx['kunta_nimi'] = dx['kunta_nimi'].str.strip()
dy['kunta_nimi'] = dy['kunta_nimi'].str.strip()
```

```python
dxdy = dy.merge(dx, on='kunta_nro', how="left", suffixes=('', '_right'))
dxdy
```

```python
dxdy.drop(['aikatyyppi','vuosikuukausi','kunta_nimi_right'],axis=1)
```

```python
new_order = ['Maakunta','kunta_nro','kunta_nimi','latitude','longitude',
             'sairauspaaryhma','ikaryhma','sukupuoli',
             'saaja_lkm','kuukausi_nro','vuosi']

dxdy = dxdy[new_order]
dxdy
```

***Puuttuvien arvojen prosenttiosuus:***

```python
(dxdy.isnull().sum()/ len(dxdy))*100
```

***Pudotetaan rivit, joissa on puutteellista tietoa.***

```python
dxdy.dropna(inplace=True)
```

```python
(dxdy.isnull().sum()/ len(dxdy))*100
```

```python
dxdy=dxdy.sort_values(by="vuosi")
```

```python
dxdy.to_csv('merge/merge1&_avoin-data_suomen_työkyvyttömyyseläkkeensaajat_2003-2023_merge.csv', index=False)
```
